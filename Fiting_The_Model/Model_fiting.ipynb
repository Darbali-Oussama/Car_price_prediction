```python
import pandas as pd
data = pd.read_csv("/kaggle/input/voitavito/avito_prix_des_voitures.csv")
data.info()
```

```python
data.shape
```

```python
data.drop(columns=data.columns[0], axis=1,  inplace=True)
data.info()
```

```python
data['Carburant'] = data['Carburant'].astype('category')
data['Boite_a_vitesse'] = data['Boite_a_vitesse'].astype('category')
data['Modele'] = data['Modele'].astype('category')
data['Kilometrage'] = data['Kilometrage'].astype('category')
```

```python
data.info()
data.head()
```

```python
for col in data.columns[1:]:
    print(col,":",data[col].nunique())
    print(data[col].value_counts())
    print('\n' + '*' * 10 + '\n')
```

```python
#one-hot encoding
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
data.columns
```

```python
data.iloc[3]
```

```python
y = data["prix"]
X = data.drop("prix", axis = 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
import pickle
filename = 'LR_model_prix_voitures_avito.sav'
pickle.dump(lr, open(filename, 'wb'))
```

```python
lr.score(X_test, y_test).round(4)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=42)
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test).round(4)
```
